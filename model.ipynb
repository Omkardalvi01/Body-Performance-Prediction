```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('bodyPerformance.csv')
df.head(10)
```

```python
df.corr(numeric_only=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
lbl_encoder = LabelEncoder()
df['class'] = lbl_encoder.fit_transform(df['class'])
```

```python
df.corr(numeric_only=True)
```

```python
df.describe()
```

```python
X = df.drop(['class'],axis=1)
Y = df['class']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, train_size = 0.8)
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot = OneHotEncoder(sparse_output=False)
one_hot_values = one_hot.fit_transform(X_train[['gender']])
one_hot_df = pd.DataFrame(one_hot_values, columns=one_hot.get_feature_names_out(['gender']))
X_train = pd.concat([X_train.reset_index(drop=True), one_hot_df.reset_index(drop=True)],axis=1)
X_train.drop(columns=['gender'],inplace=True)
one_hot_values = one_hot.transform(X_test[['gender']])
one_hot_df = pd.DataFrame(one_hot_values, columns=one_hot.get_feature_names_out(['gender']))
X_test = pd.concat([X_test.reset_index(drop=True), one_hot_df.reset_index(drop=True)],axis=1)
X_test.drop(columns=['gender'],inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
Scaler = MinMaxScaler()
X_train = Scaler.fit_transform(X_train)
X_test = Scaler.transform(X_test)
```

```python
from sklearn.svm import SVC
model = SVC(kernel='poly', degree=6)
model.fit(X_train,Y_train)
```

```python
predict = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(Y_test, predict)
print(f'Confusion Matrix: \n{cm} \n {classification_report(Y_test, predict)}')
```

```python
import pickle
with open('model.pkl','wb') as f:
    pickle.dump((model,Scaler,one_hot,lbl_encoder),f)
```
